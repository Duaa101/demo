```python
#The project done by : Fatimah alzuwayyid - Shahd Alghozwi F74


def main(): #The main function that contains all program functions in it.
    flag=True 
    menu() #Calling the menu function to print it. 
    while flag :
        userChoice=input("Enter your choice:") #Asking the user to enter the choice.
        if userChoice=="1":
            MovieID=input("Enter the Movie ID: ") #Asking the user to enter the movie ID
       
            if len(MovieID)==5 and MovieID.isdigit():
                MovieName=input("Enter the Name Of the Movie: ")#Asking the user to enter the movie name
                if flag:
                    MovieAuthor=input("Enter the Writer of the Movie: ") #Asking the user to enter the author name 
                    UserRating=float(input("How Do you Rate the Movie\n enter a value between 1 and 5 :")) #Asking the user to enter the value of rating
                    if UserRating >= 1 and  UserRating <=5 :#Making the user validation for the rating, whether its between 1 and 5
                            NumOfReviews = input('Enter The Number of Reivews :')#Asking the user to enter the number of reviews
                            if NumOfReviews.isdigit(): #Making the user validation for the number of reviews,it should contains only digits
                                Price=input('Enter The Price Of the Movie :')#Asking the user to enter the price
                                if Price.isdigit(): #Making the user validation for the price,it should contains only digits
                                    Year= input('Enter the Evaluation year :')#Asking the user to enter the evaluation year
                                    if Year.isdigit(): #Making the user validation for year,it should contains only digits
                                        
                                        Genre=input('Enter the Domain Of the Movie :')#Asking the user to enter the movie domain 
                                        if Genre.isalpha():
                                            addnewmovie(MovieID,MovieName,MovieAuthor,UserRating,NumOfReviews,Price,Year,Genre)#Calling the function that is sepcified in adding new movie, this function takes 8 argumants.
                                        else :
                                            print('You Have Entered invalid Genre') #If the movie domain is not in the correct form.
                                    else :
                                        print('You Have Entered invalid Year') #If the year is not in the correct form.
                                        
                                else :
                                    print('You Have Entered invalid Price') #If the price is not in the correct form.
                            else :
                                print('You Have Entered invalid Number of views') #If the number of reviws is not in the correct form.
                    else :
                          print('You Have Entered invalid Value for Rating ')#If the Rating value is not in the correct form.
                else :
                    print('You Have Entered invalid Name')#If the name is not in the correct form.
            else : 
                print('You Have Entered invalid ID Number')#If the movie ID is not in the correct form.
                
        if userChoice=="2":
            updateMovieInfo()
            
        if userChoice=="3":
            displayInfo()
            print()
        if userChoice=="4":
            displaySpecificInfo()
            print()
        if userChoice=="5":
            saveMovieInformationForGenre()
            print()
        if userChoice=="6":
            flag=False
            print("Goodbye!")
            

        

def menu(): #This function is printing the choices of the Movie Recommendation System.
    
    print('Movie Recommendation System')
    print('_'*40,'\n')
    print('1- Add new Movie information')
    print('2- Update Movie information')
    print('3- Display information for all Movie')
    print('4- Display specific information')
    print('5- Save specific data in new files')
    print('6- Exit')
    print('_'*40,'\n')

#This function is working only when the user choice is 1,and its main rule is to add a new movies
def addnewmovie(MovieID,MovieName,MovieAuthor,UserRating,NumOfReviews,Price,Year,Genre) :
    file=open("movie.txt","r") #Opening the movie.txt file in the reading mode, and save it to the file variable.
    allLines="" #Defining a new variable called "allLines" and its type is string
    for line in file:#For loop is used here to specify each line in the file.
        allLines=allLines+line#Adding the line to allLines variable.
    file.close()#Closing the file.
    if MovieID and Year in line : #Checking if the entered MovieID and the year , is in allLines variable or not
            print("Error: you entered a Movie ID  already has been reviewed for the same Year.")
            file.close()#Closing the file.
    else :
        file=open("movie.txt","w")#Opening the movie.txt file in the writing mode, and save it to the file variable.
        file.write(str(allLines)+"\n"+str(MovieID)+","+str(MovieName)+","+str(MovieAuthor)+","+str(UserRating)+","+str(NumOfReviews)+","+str(Price)+","+str(Year)+","+str(Genre))#Adding the previoues elements saved in allLines, and the new arguments in an appropraite format to the file.
        print("The new Movie was added successfully!")
        file.close()#Closing the file.
    print()

    

def updateMovieInfo(): #This function is working only when the user choice is 2,and its main rule is to update movie informations
    file=open("movie.txt","r") #Opening the movie.txt file in the reading mode, and save it to the file variable.
    movie_id = input("Enter MovieID: ") #Asking the user to enter the ID Of the movie we want to change its info 
    year = input("Enter Year: ") #Asking the user to enter the year Of the movie we want to change its info 
    lines = ""
    found = False
    for line in file : #Indicating to each line in the file.
        
        lst = line.split(',')#Making a list 
        if lst[0]== movie_id and lst[6]== year : #Verify whether the identification code was used for a year that already exists or not 
            found = True
            user_rating = input("\nDo you want to Update UserRating?\n Enter 1 for yes \n Enter 2 for no: ")#To confirm the process of user's movie updating.
            if user_rating == "1":
                newrate = float(input("\nEnter new UserRating: ")) 
                condition = True 
                while condition :
                    if float(newrate) >=1 and float(newrate)<=5 :#Making the user validation for the rating, whether its between 1 and 5
                        condition = False #To end the loop 

                    else : 
                        newrate = input('you should enter a number between 1 and 5 ') #If the condition is not met, the user will be asked to enter a new rate
            lst[3] = str(newrate) #to substitue by the new Rating value in the list               
            num_reviews = input("\nDo you want to Update NumOfReviews?\n Enter 1 for yes \n Enter 2 for no: ")#To confirm the process of user's movie updating.
            if num_reviews.lower() == "1":
                newreviews = int(input("\nEnter new NumOfReviews: "))
                lst[4]= str(newreviews) #to substitue by the new rivew value in the list
            price = input("Do you want to Update Price?\n Enter 1 for yes \n Enter 2 for no: ") #To confirm the process of user's movie updating.
            if price.lower() == "1":
                newprice = input("\nEnter new Price: ")
                lst[5] = newprice #to substitue by the new price in the list
            line =('%s,%s,%s,%s,%s,%s,%s,%s'%(lst[0],lst[1],lst[2],lst[3],lst[4],lst[5],lst[6],lst[7])) # To convert the list into string
           
            lines = lines + line
        else : 
            lines = lines + line
    file = open('movie.txt','w')#Opening the movie.txt file in the writing mode, and save it to the file variable.
    file.write(lines) #write all lines we saved in the movie.txt file 
    file.close() #Closing the file.
    print("Movie information updated successfully!")
    if not found :
        print('movie is not found')
    
    
    file.close() #Closing the file.
  
    
def displayInfo() : #This function is working only when the user choice is 3,and its main rule is to display information for the user 
    print('_'*40,'\n') #The menu of choices offered for the user
    print('(select only 3 of the following sub-options')
    print('1- Sort by MovieID')
    print('2- Sort by MovieName')
    print('3- Sort by Genre')
    print('4- Sort by Price')
    print('5- Sort by Year')
    print('_'*40,'\n')
    
    numbers = []
    for i in range(3): #to enter 3 different number for sub-options
        while True:
            num = input(f"Enter number {i+1} (between 1 and 5): ")
            if num.isdigit() and 1 <= int(num) <= 5 and int(num) not in numbers:
                numbers.append(int(num)) #to add the numbers entered in numbers list
                break
            else:
                print("Invalid input. Please enter a number between 1 and 5 that hasn't been entered before.")
    print('_'*40,'\n')            
    file = open('movie.txt','r')#Opening the movie.txt file in the reading mode, and save it to the file variable.
    byID = [] #to create empty list
    byMovieNames = []#to create empty list
    byGenre = []#to create empty list
    byPrice = []#to create empty list
    byYear = []#to create empty list
    for line in file :
            if len(line)>1 :
                line = line.strip()
                lst = line.split(",")
                byID.append(lst[0]) #To add all ID numbers in one list
                byMovieNames.append(lst[1]) #To add all Movie names in one list
                byGenre.append(lst[7]) #To add all movie genre in one list
                byPrice.append(lst[5])#To add all prices in one list
                byYear.append(lst[6])

    file.close()        
    information = {} #crate empty dictionary 
    information['ID']= byID #To indicate the ID in a specific key.
    information['MovieName']= byMovieNames #To indicate the movie name in a specific key.
    information['Genre']=byGenre
    information['Price']=byPrice
    information['Year']=byYear
    
            
        
    for i in numbers : #Indicating to each line in the file.
        if i==1:
            print('%-20s '%('Movie ID :' ) ,end='')
            for i in information['ID'] :
                print('%-20s'%i ,end="") #To print all items under a specific key in the dictionary in an orderly way  
        if i==2:
            print('-'*30)
            print('%-20s'%('Movie name :'),end='')
            for i in information['MovieName'] : #To print all items under a specific key in the dictionary in an orderly way
                print('%-20s'%i ,end='')
        if i==3:
            print('-'*30)
            print('%-20s'%('Genre :'),end='')
            for i in information['Genre'] :#To print all items under a specific key in the dictionary in an orderly way
                print('%-20s'%i,end='')
        if i==4:
            print('-'*30)
            print('%-20s'%('Price :'),end='')
            for i in information['Price'] : #To print all items under a specific key in the dictionary in an orderly way
                print('%-20s'%i ,end='')
                
        if i==5:
            print('-'*30)
            print('%-20s'%('Year :'),end='')
            for i in information['Year'] : #To print all items under a specific key in the dictionary in an orderly way
                print('%-20s'%i ,end='')
            
 
# 4.Displaying specific information: this feature allows the user to display Movies’ information based on some specified criteria. It has the following sub-options: (select only 4 of the following sub-options)•View the top 50 selling movies in the last 10 years (sorted based on Year and UserRating).•Print the Total Price for all moviesper Year.•Print the Average Price of all moviesper Year.•Print the Total number of all (unique) moviesper Author
def displaySpecificInfo():
    print('_'*40,'\n') #The menu of choices offered for the user
    print('1- Print the Average Price of all (unique) movies per Author')
    print('2- Print the Total Price for all movies per Year')
    print('3- Print the Average Price of all movies per Year')
    print('4- Print the Total number of all (unique) movies per Author')
    print('_'*40,'\n')
    userChoice = int(input("Enter your choice: ")) #To enter the user's choice
    if userChoice == 1:
        displayAverageNumberPerAuthor()
    elif userChoice == 2:
        displayTotalPricePerYear()
    elif userChoice == 3:
        displayAveragePricePerYear()
    elif userChoice == 4:
        displayTotalNumberPerAuthor()
    else:
        print("Invalid choice!")

def displayAverageNumberPerAuthor():
    file = open('movie.txt','r')#Opening the movie.txt file in the reading mode, and save it to the file variable.



def displayAverageNumberPerAuthor():
    # Create an empty dictionary to store the total number of movies and total reviews for each author
    author_stats = {}

    # Open the file for reading
    f = open('movie.txt','r')
        # Loop through each line in the file
    for line in f:
        # Split the line into movie attributes
        attributes = line.strip().split(',')
        # If the line is not empty
        if len(attributes) > 1:
            # Extract the movie author and number of reviews
            author = attributes[2]
            reviews = float(attributes[3])
            # If the author of the current movie is already in the dictionary, update their total number of movies and total reviews
            if author in author_stats:
                author_stats[author]['total_movies'] += 1
                author_stats[author]['total_reviews'] += reviews
            # If the author is not in the dictionary yet, add them and set their total number of movies and total reviews to the current movie's values
            else:
                author_stats[author] = {'total_movies': 1, 'total_reviews': reviews}

    # Print a header for the output
    print("Average number of movies and reviews per author")
    # Print a table header with column names for the output
    print('%-20s%-20s%-20s' % ("Author", "Average Movies", "Average Reviews"))
    # Loop through the dictionary of author movie and review stats and print each author's name and average number of movies and reviews
    for author, stats in author_stats.items():
        avg_movies = stats['total_movies'] / len(author_stats)
        avg_reviews = stats['total_reviews'] / stats['total_movies']
        print('%-20s%-20.2f%-20.2f' % (author, avg_movies, avg_reviews))
    
    
def displayTotalPricePerYear():
    # Create an empty dictionary to store the total price of movies for each year
    year_prices = {}

    # Open the file for reading
    f = open('movie.txt','r')
    # Loop through each line in the file
    for line in f:
        # Split the line into movie attributes
        attributes = line.strip().split(',')
        # If the line is not empty
        if len(attributes) > 1:
            # Extract the movie year and price
            year = int(attributes[6])
            price = int(attributes[5])
            # If the year of the current movie is already in the dictionary, update its total price
            if year in year_prices:
                year_prices[year] += price
            # If the year is not in the dictionary yet, add it and set its total price to the current movie's price
            else:
                year_prices[year] = price

    # Print a header for the output
    print("Total price of movies per year")
    # Print a table header with column names for the output
    print('%-10s%-10s' % ("Year", "Total Price"))
    # Loop through the dictionary of year movie prices and print each year and its total movie price
    for year, price in year_prices.items():
        print('%-10s%-10s' % (year, price))
        
def displayAveragePricePerYear():
    # Create an empty dictionary to store the total price and count of movies for each year
    year_stats = {}

    # Open the file for reading
    f = open('movie.txt','r')
        # Loop through each line in the file
    for line in f:
        # Split the line into movie attributes
        attributes = line.strip().split(',')
        # If the line is not empty
        if len(attributes) > 1:
            # Extract the movie year and price
            year = int(attributes[6])
            price = int(attributes[5])
            # If the year of the current movie is already in the dictionary, update its total price and count
            if year in year_stats:
                year_stats[year][0] += price
                year_stats[year][1] += 1
            # If the year is not in the dictionary yet, add it and set its total price and count to the current movie's price and 1
            else:
                year_stats[year] = [price, 1]

    # Print a header for the output
    print("Average price of movies per year")
    # Print a table header with column names for the output
    print('%-10s%-10s' % ("Year", "Average Price"))
    # Loop through the dictionary of year movie stats and print each year and its average movie price
    for year, stats in year_stats.items():
        # Calculate the average price by dividing the total price by the count
        average_price = stats[0] / stats[1]
        print('%-10s%-10.2f' % (year, average_price))
        
def displayTotalNumberPerAuthor():
    # Create an empty dictionary to store the total number of movies for each author
    author_stats = {}

    # Open the file for reading
    f = open('movie.txt','r')
    # Loop through each line in the file
    for line in f:
        # Split the line into movie attributes
        attributes = line.strip().split(',')
        # If the line is not empty
        if len(attributes) > 1:
            # Extract the movie author
            author = attributes[2]
            # If the author of the current movie is already in the dictionary, increment their total number of movies
            if author in author_stats:
                author_stats[author] += 1
            # If the author is not in the dictionary yet, add them and set their total number of movies to 1
            else:
                author_stats[author] = 1

    # Print a header for the output
    print("Total number of movies per author")
    # Print a table header with column names for the output
    print('%-20s%-10s' % ("Author", "Total Number"))
    # Loop through the dictionary of author movie counts and print each author's name and total number of movies
    for author, number in author_stats.items():
        print('%-20s%-10s' % (author, number))
    
def saveMovieInformationForGenre():
    # Open the input file for reading
    f =  open('movie.txt','r')
    # Read all the lines in the file
    lines = f.readlines()
    genre = input("Enter a genre: ")
    # check if genre exists
    # Loop through each line in the file
    for line in lines:
        # Split the line into movie attributes
        attributes = line.strip().split(',')
        # If the genre of the current movie matches the requested genre, write the line to the new file
        if attributes[7].lower() == genre.lower():
            print('Genre Found, Saving...')
            break

    # Create a new file for writing the movie information for the given genre
    genre_file_path = f"{genre}_movies.txt"
    f = open(genre_file_path,'w')
    # Loop through each line in the file
    for line in lines:
        # Split the line into movie attributes
        attributes = line.strip().split(',')
        # If the genre of the current movie matches the requested genre, write the line to the new file
        if attributes[7].lower() == genre.lower():
            f.write(line)
    
    print(f"Movie information for {genre} saved to {genre_file_path}")
        

main()


```
